```python
import numpy as np
from sklearn.metrics import average_precision_score, roc_auc_score
from sklearn.ensemble import IsolationForest
import sys
from Chains import Chains
import matplotlib.pyplot as plt
import pandas as pd

# synthetic data point classes
CLASSES = [range(0,1000), # sparse benign cluster
           range(1000,3000), # dense benign cluster
           range(3000,3050), # clustered anomalies
           range(3050, 3075), # sparse anomalies
           range(3076,3082), # local anomalies
           range(3075,3076)] # single anomaly


k = 50
nchains = 50
depth = 10

X = np.loadtxt("synDataNoisy.tsv")
y = np.array([0] * 3000 + [1] * 82)
df = pd.read_csv("synDataNoisy.tsv", sep="\t", header=None)
print(df)

cf = Chains(k=k, nchains=nchains, depth=depth)
cf.fit(X)
anomalyscores = -cf.score(X)
print("anomoly scores: ", anomalyscores)
ap = average_precision_score(y, anomalyscores)
auc = roc_auc_score(y, anomalyscores)
print("xstream: AP =", ap, "AUC =", auc)
df[104] = anomalyscores
```

```python
# https://stackoverflow.com/a/62834216
import matplotlib.lines as mlines
ax = plt.axes(projection='3d')
marker_dict = {0.0: ".", 1.0: "x"}

cmap=plt.cm.Reds
colors = cmap(np.linspace(0,1,len(df[104])))

fig = plt.figure(figsize = (10,8), dpi=150)
ax = fig.add_subplot(projection='3d')
for i in range(len(df[0])):
    g=ax.scatter(df.iloc[i, 0], df.iloc[i, 1], df.iloc[i, 2], c=colors[i], marker=marker_dict[df.iloc[i, 103]])


norm = plt.Normalize(np.min(df[104]), np.max(df[104]))
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
ax.figure.colorbar(sm, label="xStream Score")

benign = mlines.Line2D([], [], color='k', marker='.', linestyle='None',
                          markersize=10, label='Benign')
anomalies = mlines.Line2D([], [], color='k', marker='x', linestyle='None',
                          markersize=10, label='Anomaly')

plt.legend(handles=[benign, anomalies])

plt.show()
```
